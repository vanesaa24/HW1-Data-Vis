```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
Air_pollution = pd.read_csv("global_air_pollution_dataset.csv")
Lung_Cancer = pd.read_csv("lung_cancer_prediction_dataset.csv")
```

# Part 3 
## Task 1. Create a Boxplot of Lung Cancer Deaths Distribution.

```python
sns.boxplot(x='Annual_Lung_Cancer_Deaths', data=Lung_Cancer, showmeans=True)
plt.title('Lung Cancer Deaths Distribution')
plt.xlabel('Deaths')
plt.show()
```

## Task 2. Create a Histogram of PM2.5 AQI Values.

```python
sns.histplot(data=Air_pollution, x='PM2.5_AQI_Value')
plt.title('Histogram of PM2.5 AQI Values')
plt.xlabel('PM2.5 AQI Value')
plt.ylabel('Count')
plt.show()
```

## Task 3. Create a Density Plot of the Lung Cancer Mortality Rate.

```python
sns.kdeplot(data=Lung_Cancer, x='Mortality_Rate')
plt.title('Density Plot of the Lung Cancer Mortality Rate')
plt.xlabel('Mortality Rate')
plt.ylabel('Density')
plt.show()
```

# Part 4.
## Task 1.

```python
plt.figure(figsize=(10, 6))

sns.histplot(Air_pollution['PM2.5_AQI_Value'], bins=50, kde=False,
             stat='density', color='slateblue', edgecolor='black', 
             alpha=0.7, label='Histogram', linewidth=0.3)
sns.kdeplot(Air_pollution['PM2.5_AQI_Value'], color='darkred', 
            linewidth=2, label='Density Plot (KDE)', fill = True)
plt.title('PM2.5 AQI Distribution with Density Overlay', fontsize=14)
plt.xlabel('PM2.5 AQI Value', fontsize=12)
plt.ylabel('Density', fontsize=12)  
plt.xlim(-20, 520)
plt.grid(axis='y', alpha=0.3)
plt.legend(fontsize=10)
plt.figtext(0.5, 0.01, "This plot represents the distribution of PM2.5 AQI values with a density overlay.", 
            ha="right", fontsize=10)
plt.tight_layout(rect=[0, 0.05, 1, 1])
plt.show()
```
